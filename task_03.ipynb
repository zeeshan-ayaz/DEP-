# Data Analysis Project: Titanic Dataset

### Introduction
In this project, we'll explore and visualize the Titanic dataset using pandas and matplotlib. This dataset provides information about passengers and their survival status, along with various attributes like age, class, and fare.

### 1. Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### 2. Load Dataset

```python
# Load the dataset from a CSV file
data = sns.load_dataset("titanic")

# Display the first few rows of the dataset
data.head()
```

### 3. Data Overview

```python
# Display basic information about the dataset
data.info()

# Display summary statistics
data.describe(include='all')
```

### 4. Data Cleaning

```python
# Check for missing values
missing_values = data.isnull().sum()

# Drop rows with missing target values (survived) and fill or drop other missing values as needed
data_clean = data.dropna(subset=['survived'])
data_clean = data_clean.fillna({'age': data_clean['age'].median(), 'embarked': 'S'})

# Display the cleaned dataset
data_clean.head()
```

### 5. Exploratory Data Analysis (EDA)
#### 5.1 Distribution of Happiness Scores

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='survived', data=data_clean)
plt.title('Distribution of Survival Status')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.xticks([0, 1], ['No', 'Yes'])
plt.show()
```

#### 5.2 Survival Rate by Class

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='pclass', y='survived', data=data_clean, ci=None)
plt.title('Survival Rate by Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```

### 5.3 Age Distribution by Survival Status

```python
plt.figure(figsize=(10, 6))
sns.histplot(data_clean[data_clean['survived'] == 1]['age'], bins=30, kde=True, label='Survived', color='green')
sns.histplot(data_clean[data_clean['survived'] == 0]['age'], bins=30, kde=True, label='Not Survived', color='red')
plt.title('Age Distribution by Survival Status')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

### 5.4 Fare Distribution by Survival Status

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='survived', y='fare', data=data_clean)
plt.title('Fare Distribution by Survival Status')
plt.xlabel('Survived')
plt.ylabel('Fare')
plt.xticks([0, 1], ['Not Survived', 'Survived'])
plt.show()
```

## 6. Insights and Conclusion

### Insights
1. **Distribution of Survival Status**: The dataset shows a significant difference in the number of passengers who survived versus those who did not.
2. **Survival Rate by Class**: The survival rate varies by passenger class, with higher survival rates generally observed in first class.
3. **Age Distribution by Survival Status**: The age distribution differs between survivors and non-survivors, with different patterns of survival across age groups.
4. **Fare Distribution by Survival Status**: Fare distribution varies between those who survived and those who did not, with survivors generally having a higher median fare.

## Conclusion
This analysis provides insights into the factors influencing survival on the Titanic. The findings can help in understanding how different attributes such as class, age, and fare impacted the likelihood of survival.
